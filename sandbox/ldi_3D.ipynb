```python
import lindelint as ldi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
points = np.array([[-7.2, -1.3, -6.0],
                   [ 5.1,  6.3, -7.6],
                   [-2.4, -1.3,  0.2],
                   [ 8.9,  4.9,  5.1],
                   [-5.4, -6.3,  7.2],
                  ])

properties = np.array([[ 2.5,  0.5, -0.2],
                       [ 0.0, -1.3, -1.6],
                       [ 0.5,  1.4,  2.8],
                       [ 2.3,  1.4,  0.8],
                       [-0.8,  1.8,  1.5]])
```

```python
points = np.array([[-5, 0, 0],
                   [ 5, 0, 0],
                   [ 0, 7, 0],
                   [ 0, 0, 7],
                  ])

properties = np.array([[ 2.5,  0.5, -0.2],
                       [ 0.0, -1.3, -1.6],
                       [ 0.5,  1.4,  2.8],
                       [-0.8,  1.8,  1.5]])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(points[:,0], points[:,1], points[:,2],
          properties[:,0], properties[:,1], properties[:,2],
          color='r')

ax.axes.set_xlim3d(-10, 10) 
ax.axes.set_ylim3d(-10, 10) 
ax.axes.set_zlim3d(-10, 10) 

plt.show()
```

```python
new_points = np.random.uniform(-10, 10, (200, 3))
```

```python
interpolator = ldi.Interpolator(points, properties)
```

```python
new_properties = interpolator.do_your_thing(new_points)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(points[:,0], points[:,1], points[:,2],
          properties[:,0], properties[:,1], properties[:,2],
          color='r')

ax.quiver(new_points[:,0], new_points[:,1], new_points[:,2],
          new_properties[:,0], new_properties[:,1], new_properties[:,2],
          color='k')


ax.axes.set_xlim3d(-10, 10) 
ax.axes.set_ylim3d(-10, 10) 
ax.axes.set_zlim3d(-10, 10) 

plt.show()
```
